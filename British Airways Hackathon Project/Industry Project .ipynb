```python
custom_palette = ['#CE220F','#2D5C99']
fig, axs = plt.subplots(7, 2, figsize=(20, 4*len(american_df.columns[7:-3])))
for i, col in enumerate(american_df.columns[7:-3]):
    row, col_idx = i // 2, i % 2
    plot_data = american_df.groupby('Class')[col].mean().reset_index()
    ax = axs[row, col_idx]
    sns.barplot(data=plot_data, x='Class', y=col, hue='Class', ax=ax, palette=custom_palette, dodge=False)
    ax.tick_params(axis='x', labelrotation=90)
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['Business', 'Eco'])
    ax.set_xlabel('Class')
    ax.set_ylabel(col)
    ax.set_title(f'Average Rating for {col} in Business vs Economy Class')
    fig.tight_layout()
```

```python
# Create a dictionary to map class to colors
class_color_map = {'Business': '#CE220F', 'Eco': '#2D5C99'}
# Create a list of review columns to iterate through
review_columns = american_df.columns[7:-3]
# Create a for loop to iterate through each review column
for column in review_columns:
    # Group the data by age and class and calculate the mean review score
    grouped_data = american_df.groupby(['age_bin', 'Class'])[column].mean().reset_index()
    # Use seaborn to create a bar plot of the mean review score
    plot = sns.barplot(x='age_bin', y=column, hue='Class', data=grouped_data, palette=class_color_map)
    # Set the title of the plot to the name of the review column
    plot.set_title(column)
    # Show the plot
    plt.show()
```
